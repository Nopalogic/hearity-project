#### **Take a random sample to visualize**

Choose the sample randomly, not based on the order

```python
sample = data[data.patient_id == 50]
sample = sample[['left_freq_500_hz',
                 'left_freq_1000_hz',
                 'left_freq_2000_hz',
                 'left_freq_4000_hz',
                 'right_freq_500_hz',
                 'right_freq_1000_hz',
                 'right_freq_2000_hz',
                 'right_freq_4000_hz']]
```

display the historical data from first test till the latest test in each frequencies

```python
# define x and y axis
test = ['1st test', '2nd test', '3rd test', '4th test'] 
y_limit = (-10, 43) 

fig, axes = plt.subplots(2, 4, figsize=(16, 8))
fig.suptitle("Historical Audiometry Data in Each Frequencies", fontsize=16)
fig.tight_layout(pad=3.0, h_pad=3.0, w_pad=3.0)

# list to separate between left and right ear
left_ear_columns = ['left_freq_500_hz', 'left_freq_1000_hz', 'left_freq_2000_hz', 'left_freq_4000_hz']
right_ear_columns = ['right_freq_500_hz', 'right_freq_1000_hz', 'right_freq_2000_hz', 'right_freq_4000_hz']

for i, col in enumerate(left_ear_columns):
    axes[0, i].plot(test, sample[col].values, marker='x', color='b')
    axes[0, i].set_title(f"Left Ear - {col.split('_')[-2]} Hz")
    axes[0, i].set_ylim(y_limit)
    axes[0, i].set_ylabel("Hearing Level (dB)")
    axes[0, i].invert_yaxis()


for i, col in enumerate(right_ear_columns):
    axes[1, i].plot(test, sample[col].values, marker='o', color='r')
    axes[1, i].set_title(f"Right Ear - {col.split('_')[-2]} Hz")
    axes[1, i].set_ylim(y_limit)
    axes[1, i].set_ylabel("Hearing Level (dB)")
    axes[1, i].invert_yaxis() 

plt.show()
```

```python
left_ear_freq = [(500, latest_test['left_freq_500_hz']), 
                 (1000, latest_test['left_freq_1000_hz']), 
                 (2000, latest_test['left_freq_2000_hz']), 
                 (4000, latest_test['left_freq_4000_hz'])]

right_ear_freq = [(500, latest_test['right_freq_500_hz']), 
                  (1000, latest_test['right_freq_1000_hz']), 
                  (2000, latest_test['right_freq_2000_hz']), 
                  (4000, latest_test['right_freq_4000_hz'])]

# unzip the data points into x and y coordinates
x_values_1, y_values_1 = zip(*left_ear_freq)
x_values_2, y_values_2 = zip(*right_ear_freq)

# set y axis
y_limit = (-10, 70)

fig, ax = plt.subplots(1, 2, figsize=(15, 6))
ax[0].plot(x_values_1, y_values_1, color='blue', linestyle='-', linewidth=2, label='Left Ear', marker='x')
ax[0].scatter(x_values_1, y_values_1, facecolor='none', s=100, linewidth=2)
ax[1].plot(x_values_2, y_values_2, color='red', linestyle='-', linewidth=2, label='Right Ear')
ax[1].scatter(x_values_2, y_values_2, edgecolor='red', facecolor='none', s=100, linewidth=2)

for axis in ax:
    axis.set_xticks([500, 1000, 2000, 4000])
    axis.set_xticklabels(['500', '1k', '2k', '4k'])
    axis.set_ylim(y_limit)
    axis.set_xlabel('Frequency (Hz)')
    axis.set_ylabel('Hearing Loss (dB)')
    axis.invert_yaxis()
    axis.grid(True)

ax[0].set_title('Left Ear Hearing Test Result')
ax[1].set_title('Right Ear Hearing Test Result')
ax[0].legend()
ax[1].legend()
plt.tight_layout()
plt.show()
```

### **Summary of Visualizations**

**Gradual Decline in Hearing**

From the first to the fourth test, there has been a significant and gradual decline in hearing sensitivity across multiple frequencies. This consistent decline over time suggests a progressive worsening of hearing sensitivity, particularly at higher frequencies.


**Frequency-Specific Decline**

Audiometry results indicate a notable increase in hearing threshold at 4 kHz, while the hearing thresholds at lower frequencies, such as 500 Hz, 1 kHz, and 2 kHz, remain constant or show a slight decrease at first. This pattern suggests that the most pronounced decline in hearing sensitivity occurs at higher frequencies, especially at 4 kHz, where thresholds are the highest.


**Potential for Noise-Induced Hearing Loss**

The gradual decline in hearing sensitivity, particularly at higher frequencies such as 3 kHz and 4 kHz, suggests the potential for noise-induced hearing loss. This type of loss typically affects higher frequencies first, while lower frequencies may remain within normal hearing thresholds. However, the elevated thresholds at higher frequencies are concerning for long-term auditory health.


**Linear Trend in Hearing Decline**

The graph reveals a clear linear trend in hearing thresholds over time, indicating a steady increase in hearing level with each successive test. Importantly, this trend shows no signs of seasonality, as it is unaffected by monthly patterns, seasonal changes, or external events. This lack of cyclical variation suggests that the decline is due to consistent, prolonged exposureâ€”potentially from occupational noise or frequent use of earbudsâ€”rather than external, time-bound factors.

**Data Characteristic**

For each frequency (likely referring to audiometry test results at different sound frequencies), the lowest value recorded is 0. This is expected and normal because 0 could represent a response. However, the maximum value can reach up to 50, indicating the highest level of response recorded for that frequency. It's a natural variation, and there's nothing wrong with this rangeâ€”itâ€™s simply a characteristic of the data.

## **Summary**

**Stationarity Status**

The data is non-stationary due to the presence of a clear trend, which indicates a consistent increase in hearing level over time rather than fluctuating around a constant mean. Because these data are non-stationary, it means it has a shifting mean and potentially changing variance over time, both of which are typical in data with strong trends. 

**Feature Engineering**


* Not every column in the dataset is relevant to the analysis. Feature selection is the process of choosing the most important features (or columns) that will be used in the analysis or model building. By selecting only the relevant features, it can improve the performance of the model, make the model easier to interpret and feature selection could help by eliminating unnecessary columns and focusing on the most important data.

* The data only contains 4 data points, which represent tests conducted on specific dates within a two-year period, from January 15, 2022 to July 27, 2023. These tests are conducted every six months. To enhance the dataset and provide more granular information, resampling is required. Specifically, the resampling method is upsampling to increase the frequency of the data. This means that instead of having only one data point for each six-month period, the preprocessing step before modelling will generate new data points that represent daily values. The goal is to interpolate or create additional daily data between the existing six-monthly data points to enrich the dataset for more detailed analysis.

* Since the data only provides the date, having a column that represents the number of days since a certain reference date could be helpful for the model to capture any trends or patterns over time.

* Data needs to be scaled because the values across different frequencies in the hearing test vary significantly. Without scaling, certain frequencies with larger ranges may dominate the analysis, leading to biased results. Scaling ensures that all frequencies contribute equally, improving the performance and fairness of any predictive model or analysis.
